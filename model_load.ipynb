```python
from roboflow import Roboflow
from google.colab import files
from ultralytics import YOLO
```

```python
api_key = "un_api_key_de_roboflow"

rf = Roboflow(api_key=api_key)
project = rf.workspace("aaron-qwuzu").project("pokemon-cards-63wlp")
version = project.version(5)
version.download("yolov8")
```

```python
!nvidia-smi
```

```python
# Entrenamiento
model = YOLO("yolov8n.pt")

model.train(
    data="Pokemon-cards-5/data.yaml",
    epochs=50,
    imgsz=640,
    project="pokemon_output",
    name="train"
)

print(" Entrenamiento terminado")
print(" Pesos guardados en: pokemon_output/train/weights/best.pt")
```
