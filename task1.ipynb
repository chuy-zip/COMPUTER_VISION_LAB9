- por que U-Net es ideal para separar humanos del fondo, pero falla al distinguir objetos iguales superpuestos;
- por que Mask R-CNN si puede separar e iluminar diferentes armas, incluso cuando se tocan en la imagen.

## 1. Diferencia : Segmentacion Semantica vs Segmentacion de Instancia

La segmentación semántica opera a nivel de clase, asignando una etiqueta a cada pixel de la imagen, resultando en  un mapa de clases en donde todos los objetos de la misma categoría son indistinguibles entre sí. 

La segmentación de instancia opera a nivel de objeto, distingue entre instancias separadas de la misma clase y cada instancia tiene una máscara binaria.

La diferencia entonces se encuentra a nivel de escala, la segmentación semántica se preocupa por responder, 

**¿A que clase corresponde este pixel?**

Cuando la segmentación de instancia responde 

**¿Quien soy dentro de $X$ clase?**

## 2. El Caso U-Net

### 2.1 Por que U-Net es ideal para separar humanos del fondo
U-net contrae la imagen preguntando,

**¿Hay una persona aquí?** 

Esto a través de la arquitectura encoder-decoder, el encoder captura el contexto semántico, y el decoder reconstruye la resolución espacial píxel a píxel usando skip connections que recuperan detalles finos. 

Además es computacionalmente eficiente y solo tiene dos clases de salida. Por un lado, U-net no requiere propuestas de regon ni etapas de clasificación, en una sola pasada se produce el mapa de segmentación.




### 2.2 Por que U-Net fracasaria en la tarea de las espadas cruzadas

Para la Tarea 2, al producir un único mapa de segmentación, si existen dos objetos de una misma clase sobrepuestos, como por ejemplo dos espadas, se genera la misma clase para las dos espadas. Dicho de otra mánera, el modelo solo responde a que clase corresponde la espada, si hay dos o más espadas.

## 3. El Caso Mask R-CNN

Mask R-CNN resuelve este problema porque es un modelo de **segmentacion de instancia** basado en una arquitectura de **dos etapas**.

### 3.1 Primera etapa: propuesta de regiones
 La red genera bounding boxes candidatas ,
 **hay algo de interés aquí**
 
  Cada propuesta es una región de la imagen independiente. Aquí ya se produce una separación espacial de objetos, incluso si se tocan.

  

### 3.2 Segunda etapa: clasificacion, caja y mascara por instancia
Clasificación + Máscara por instancia: Para cada propuesta de región, la red ejecuta simultáneamente 
- Clasificación de clase
- Refinamiento del bounding box
- Predicción de una máscara binaria dentro de esa región específica.

 La máscara se predice de forma completamente aislada para cada instancia.

### 3.3 Por que Mask R-CNN si puede iluminar una espada de rojo y otra de azul

En la primera etapa se logra detectar dos objetos separados  y genera dos bounding boxes diferentes, sin importar que se solapen. Luego, cada bounding box se procesa de mánera independiente y se predice una máscara para la espada dentro de cada bounding box.

Al tener una ID de instancia para cada máscara, no importa que dos píxeles sean adyacentes o se superpongan.

Mask R-CNN razona primero sobre objetos como entidades  y luego segmenta dentro de cada entidad. U-Net razona directamente sobre píxeles como clases sin noción de individualidad de objeto.
